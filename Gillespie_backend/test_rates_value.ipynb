```python
import numpy as np
import matplotlib.pyplot as plt
ell_tot = 100.
distance_anchor = 50.
rho0 = 0.01
temperature = 1.
```

```python
def unbinding_rate(temperature):
    return np.exp(-1./temperature)#/(1+np.exp(-1/temperature))
def omega(D,ell):
    return 3/(2*np.arccos(-1)*ell)**1.5*np.exp(-3/2*D**2/ell)
def binding_rate(DR1,DR2,ell1,ell2):
    print(omega(DR1,ell1))
    print(omega(DR2,ell2))
    print(omega(DR1+DR2,ell1+ell2))
    return omega(DR1,ell1)*omega(DR2,ell2)/(omega(DR1+DR2,ell1+ell2)+omega(DR1,ell1)*omega(DR2,ell2))
```

```python
temp = np.linspace(0.001,10,10000)
plt.plot(temp,unbinding_rate(temp))
```
